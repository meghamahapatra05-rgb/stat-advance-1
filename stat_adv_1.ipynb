1. What is random variable in probability theory?

3. Explain the difference between discrete and continuous distribution.
-> A discrete distribution deals with countable outcomes and uses probability mass function while a continuous distribution deals with uncountable values and uses probability density function where probability at single point is zero.

4. What is a binomial distribution, and how is it used in probability?
-> A binomial distribution models is the number of successes in a fixed number of independent trials, each with same probability of success. It is used when outcomes are only success or failures.

5. What is the standard normal distribution and why is it important?

6.What is the central limit theorem, and why is it critical in statistics?

7. What is the significance of confidence intervals in statistical analysis?
-> Confidence intervals provide a range of values within which the true population parameter is likely to lie with a certain confidence level. They measure the raliability and accuracy of sample estimates.

8.What is the concept of expected value in a probility distribution?

9. Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute it's mean and standard deviation using Numpy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 5, 1000)
mean = np.mean(data)
std = np.std(data)
print("Mean:", mean)
print("Standard Deviation:", std)
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Normal Distribution')
plt.show()
```

10. You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260] -> Explain how would you apply Central Limit Theorem to estimate the average sales with 95% confidence interval. -> Write a python code to compute the mean sales and its confidence interval.

```python
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255, 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

confidence_interval = stats.norm.interval(0.95, loc=mean_sales, scale=std_sales/np.sqrt(n))
print("Mean sales:", mean_sales)
print("95% confidence interval:", confidence_interval)
```
